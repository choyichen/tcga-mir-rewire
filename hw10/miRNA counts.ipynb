```python
md = pd.read_table("miRNAdisruptedcounts.txt", sep="\t")
md = md.set_index("Unnamed: 0")
md.index.name = None
md = md.dropna(how="all")
md.to_csv("miR_disrupted_counts.txt", sep="\t")
```

```python
md2 = md / md.sum()
md2 = md2.dropna(how="all")
md2.to_csv("miR_disrupted_counts.normalized.txt", sep="\t")
```

```python
mc = pd.read_table("miRNAcreatedcounts.txt", sep="\t")
mc = mc.set_index("Unnamed: 0")
mc.index.name = None
mc = mc.dropna(how="all")
mc.to_csv("miR_created_counts.txt", sep="\t")
```
