```python
import networkx as nx
import pandas as pd
from pylab import *
%matplotlib inline
import numpy as np
```

```python
G = nx.read_edgelist("C:/Users/jtso/My Documents/Cancer miRNA rewiring project/targetmirnas2.txt", delimiter="\t", create_using=nx.DiGraph())
G.remove_edge("miRNA", "Genesymbol")  # remove the header line
```

```python
hey = 0
colors2 = []
miR = "hsa-miR-548c-3p"
targets = G.neighbors(miR)
miR3 = "hsa-miR-548aj-3p"
targets3 = G.neighbors(miR3)
miR4 = "hsa-miR-548aq-3p"
targets4 = G.neighbors(miR4)
miR5 = "hsa-miR-548z"
targets5 = G.neighbors(miR5)
mirs = [miR,miR3,miR4,miR5]
fhb = fha[fha["miRNA"].isin(mirs)]
g = G.subgraph(mirs + targets + targets3 + targets4 + targets5)
for j,k in g.edges():
    test1 = []
    hey += 1
    print hey, '\r',
    for idx,row in fhb.iterrows():
        if not row["miRNA"] == j:
            continue
        if not row["Genesymbol"] == k:
            continue
        if row["Genesymbol"] in test1:
            continue
        test1.append(row["Genesymbol"])
        if row["Targetsiteclass"] == "Created":
            colors2.append("g")
        else:
            colors2.append("r")
colors = ['orange' if n in mirs else 'b' for n in g.nodes()]
figure(figsize=(20,20))
nx.draw(g, node_color=colors, alpha=.2, with_labels=True, edge_color=colors2)
```

```python
bgenes = ['AKT1', 'BCL11A', 'PTPRB', 'PBX1', 'MDM4', 'CRTC1', 'NONO']
pgenes = ['AKT1','BCL11A','NIN','SPECC1','NCOR1','MDM2','PTPRB','PBX1','MDM4','CRTC1','NONO']
hey = 0
colors2 = []
colors = []
shape = []
miR = "hsa-miR-548c-3p"
targets = G.neighbors(miR)
miR3 = "hsa-miR-548aj-3p"
targets3 = G.neighbors(miR3)
miR4 = "hsa-miR-548aq-3p"
targets4 = G.neighbors(miR4)
miR5 = "hsa-miR-548z"
targets5 = G.neighbors(miR5)
mirs = [miR,miR3,miR4,miR5]
fhb = fha[fha["miRNA"].isin(mirs)]
g = G.subgraph(mirs + targets + targets3 + targets4 + targets5)
pos = nx.spring_layout(g)
for j,k in g.edges():
    test1 = []
    hey += 1
    print hey, '\r',
    for idx,row in fhb.iterrows():
        if not row["miRNA"] == j:
            continue
        if not row["Genesymbol"] == k:
            continue
        if row["Genesymbol"] in test1:
            continue
        test1.append(row["Genesymbol"])
        if row["Genesymbol"] in (pgenes + tspgenes):
            if row["Targetsiteclass"] == "Created":
                colors2.append("g")
            elif row["Targetsiteclass"] == "Disrupted":
                colors2.append("r")
        else:
            colors2.append("silver")

ugenes= ['SPECC1', 'MDM2', "NIN", "NCOR1"]
tsugenes= ['RND3','RECK','EGR1']
rgenes = []
for a in g.nodes():
    if not a in (mirs + pgenes + tspgenes):
        rgenes.append(a)
d={}
for i in (mirs + pgenes + tspgenes):
    d[i] = i
figure(figsize=(20,20))
nx.draw_networkx_nodes(g, pos, nodelist=rgenes, node_color = "b", alpha = .3, node_shape = "o")
nx.draw_networkx_nodes(g, pos, nodelist= mirs, node_color = "orange", alpha=.3, node_shape = "^")
nx.draw_networkx_nodes(g, pos, nodelist= bgenes, node_color = "lime", alpha=.3, node_shape = "s")
nx.draw_networkx_nodes(g, pos, nodelist = ugenes, node_color = "fuchsia", alpha=.3, node_shape = "s")
nx.draw_networkx_nodes(g, pos, nodelist= tsbgenes, node_color = "lime", alpha=.3, node_shape = "d")
nx.draw_networkx_nodes(g, pos, nodelist = tsugenes, node_color = "fuchsia", alpha=.3, node_shape = "d")
nx.draw_networkx_edges(g, pos, edge_color = colors2, alpha = .4)
nx.draw_networkx_labels(g, pos, labels = d)
```
