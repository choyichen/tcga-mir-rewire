```python
import pandas as pd
```

```python
fh = pd.read_table("C:/Users/jtso/My Documents/Cancer miRNA rewiring project/oncogene_target_somatic.txt", sep="\t")
fh.head()
```

```python
fh = fh[["Genesymbol", "miRNA", "Cancertype"]]
fh.head()
```

```python
type2df = {}
types = ["Acute myeloid leukemia", "BladderUrothelialCarinoma(TCGA,US)", "Breast Cancer", "BreastCarcinoma(WTSI,UK)", "BreastInvasiveCarcinoma(TCGA,US)", "CervicalSquamousCellCarcinoma(TCGA,US)", "Chronic lymphocytic leukaemia", "ColonAdenocarcinoma(TCGA,US)", "Colorectal adenocarcinoma", "GastricCancer(CCGC,CN)", "Hepatocellular carcinoma", "Hypodiploid_ALL", "Leukemia/T-ALL", "LiverCancer(NCC_JP)", "LiverCancer(RIKEN,JP)", "Lung", "MalignantMelanoma(WTSI,UK)", "Medulloblastoma", "Melanoma", "Myeloid leukemia", "Myeloma", "Pancreatic", "PediatricBrainTumors(DKFZ,DE)", "Prostate", "ProstateCancer(OICR,CA)", "RectumAdenocarcinoma(TCGA,US)", "Small cell lung cancer", "T-cell ALL", "UterineCorpusEndometroidCarcinoma(TCGA,US)"]
for ct in types:
    type2df[ct] = fh[fh["Cancertype"] == ct]
```

```python
type2df["Acute myeloid leukemia"]
```

```python
from collections import Counter
type2dict = {}
for c,d in type2df.items():
    type2dict[c] = Counter()
    for idx, row in type2df[c].iterrows():
        type2dict[c][row["Genesymbol"]] += 1
```

```python
type2dict["Acute myeloid leukemia"]
```

```python
type2series = {}
for e,f in type2dict.items():
    values = []
    indexes = []
    for g,h in type2dict[e].items():
        values.append(h)
        indexes.append(g)
    type2series[e] = pd.Series(values, index = indexes)
    
df = pd.DataFrame(type2series)
df
```

```python
df = df.fillna(0)
df
```

```python
df.to_csv("gene_heat_matrix2.txt", sep="\t")
```
